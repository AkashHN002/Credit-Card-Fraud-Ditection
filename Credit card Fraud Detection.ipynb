```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("C:/Users/AKASH HN/Downloads/Dataset/creditcard.csv")
data.head()
```

```python
data.describe()
```

```python
data['Class'].value_counts().plot.bar(rot = 0)
plt.title('Distribution of Classes')
plt.show()
```

```python
corr = data.corr()
plt.figure(figsize = (20, 18))
sns.heatmap(corr, annot = True, cmap = 'coolwarm', fmt = '.2f')
```

```python
X = data.drop('Class', axis = 1)
y = data['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
model_dict = dict()
```

```python
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
pred =  log_model.predict(X_test)
test_accuracy = accuracy_score(y_test, pred)
model_dict['Logistic Regression'] = [test_accuracy]
print(f'Test accuracy of Logistic Regression : {test_accuracy:.2%}')
```

```python
DTC = DecisionTreeClassifier(criterion = 'entropy', max_depth = 10)
DTC.fit(X_train, y_train)
pred = DTC.predict(X_test)
test_accuracy = accuracy_score(y_test, pred)
print(f'Test accuracy of Decision Tree Classifier: {test_accuracy:.2%}')
model_dict['Decision Tree Classifier'] = [test_accuracy]
```

```python
clf = RandomForestClassifier(min_samples_split = 5, n_estimators = 20, max_depth = 10)
clf.fit(X_train, y_train)
```

```python
x_pred = clf.predict(X_train)
train_accuracy = accuracy_score(y_train, x_pred)
print(f'Train accuracy of Random forest classifier: {train_accuracy:.2%}')
y_pred = clf.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)
print(f'Test accuracy of Random forest classifier: {test_accuracy:.2%}')
model_dict['Random Forest'] = [test_accuracy]
```

```python
model_df = pd.DataFrame(model_dict).T
model_df.columns = ['Accuracy']
model_df.sort_values(by = 'Accuracy', ascending = False)
```

```python
y_pred = clf.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
disp = metrics.ConfusionMatrixDisplay(cm, display_labels = clf.classes_)
disp.plot(cmap = plt.cm.Blues, ax = plt.gca())
plt.show()
```
